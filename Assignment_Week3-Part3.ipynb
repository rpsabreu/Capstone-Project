```python
df = table[0]
df2 = df[df.Borough != 'Not assigned']
df2.head()
```

### Combined rows with same postcode but different neighbourhoods

```python
df3 = df2.groupby(['Postcode','Borough'])['Neighbourhood'].apply(', '.join).reset_index()
df3
```
